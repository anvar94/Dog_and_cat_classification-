```python
NAME = "Cats-vs-dogs-CNN"
import os
tensorboard = TensorBoard(log_dir="logsM")

pickle_in = open("X.pickle3","rb")
X = pickle.load(pickle_in)
X  = np.array(X)
pickle_in = open("y.pickle3","rb")
y = pickle.load(pickle_in)
y = np.array(y)
X = X/255.0 
```

```python
X.shape
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet(input_shape=(50, 50, 3), include_top=False)
```

```python
mobile.summary()
```

```python
model = Sequential()
for layer in mobile.layers[:-1]:
    model.add(layer)
```

```python
for layer in model.layers:
    layer.trainable = False
    
```

```python
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
type(mobile)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X, y,
          batch_size=32,
          epochs=10,
          validation_split=0.3)
```
